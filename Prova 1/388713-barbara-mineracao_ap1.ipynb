### Alternativa c)

```python
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
```

```python
X = df.values[:, :-1]
y = df.values[:, -1]
```

```python
df.shape
```

```python
X.shape, y.shape
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=42, stratify=y)
```

```python
X_train.shape, y_train.shape, X_test.shape, y_test.shape
```

```python
scaler = StandardScaler()

scaler.fit(X_train)

X_train_scl = scaler.transform(X_train)
X_test_scl = scaler.transform(X_test)
```

#### Resultado do processo de padronização

```python
#Gráfico de X_train (não padronizado)
plt.plot(X_train);
```

```python
#Gráfico de X_train (padronizado)
plt.plot(X_train_scl);
```

```python
#Gráfico de X_test (não padronizado)
plt.plot(X_test);
```

```python
#Gráfico de X_test (padronizado)
plt.plot(X_test_scl);
```

## 3ª Questão

```python
#from sklearn.metrics import accuracy_score

#Função que calcula a acurácia 
def accuracy_score(y_real, y_pred):
    return np.sum(y_pred == y_real)/y_real.shape[0]
```

### Alternativa a)

```python
#Usando conjunto de treino e conjunto de teste não padronizado (DTC)
accDTC = {'Treino não padronizado: ' : accuracy_score(y_train, y_pred_dt), 
          'Teste não padronizado: ' : accuracy_score(y_test, y_pred_dt_test)}
```

```python
#Usando conjunto de treino e conjunto de teste não padronizado (KNN)
accKNN = {'Treino não padronizado: ' : accuracy_score(y_train, y_pred_knn), 
          'Teste não padronizado: ' : accuracy_score(y_test, y_pred_knn_test)}
```

## 4ª Questão

```python
from sklearn.cluster import KMeans 
```

O valor de k pode ser escolhido utilizando a técnica do cotovelo:

```python
inertias = []

for k in range(1, 10, 2):
    model = KMeans(n_clusters=k)
    model.fit(X)
    inertias.append(model.inertia_)
```

```python
plt.plot(inertias, range(1, 10, 2), marker='o');
```
